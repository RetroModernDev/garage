### Series 

1-dimensional data type

```python
import pandas as pd

producers = pd.Series(["Commodore", "Atari", "Nintendo"])
print(producers)
```

```python
products = pd.Series(["C64", "ST", "Game Boy"])
print(products)
```

### Data frame

2-dimensional data type

```python
all_data = pd.DataFrame({"producers": producers, "products": products})
print(all_data)
```
